```python
import pandas as pd
import numpy as np
filename= '/home/kamran/stuff/coursera/Data Analysis in python/data/kc_house_data_NaN.csv'
df = pd.read_csv(filename)
df.head()
```

```python
df.drop(['id','Unnamed: 0'], axis=1, inplace=True)
df.describe()
```

```python
df['bedrooms'].isnull().sum()
```

```python
# replacing missing values with means of bedrooms
mean = df['bedrooms'].mean()
df['bedrooms'].replace(np.nan,mean, inplace=True)
df['bedrooms'].isnull().sum()
```

```python
#  replacing missing valuesin column 'bathrooms' with min-max method
minimum = df['bathrooms'].min()
maximum = df['bathrooms'].max()
modee = df['bathrooms'].mean()
# df['bathrooms'] = (df['bathrooms'] - df['bathrooms'].mean())/df['bathrooms'].std()
# min_max=(df['bathrooms']-minimum)/(maximum-minimum)
# df['bathrooms']=min_max
df['bathrooms'].replace(np.nan,modee,inplace=True)
df['bathrooms'].isnull().sum()
df.head(10)
```

```python
df['view']=df['view'].astype('int64')
df.info()
```

```python
bins = np.linspace(df['price'].min(),df['price'].max(),6)
group_names =['lower','lower_middle','middle','upper_middle','upper']
df['price_binned'] = pd.cut(df['price'],bins,labels=group_names,include_lowest=True)
df.head(10)
```
